```
# STEP 1: Import Libraries
import pandas as pd
import numpy as np

# STEP 2: Load Dataset
df = pd.read_csv('netflix_titles.csv')

# STEP 3: Basic Info
print("Initial Shape:", df.shape)
df.info()
df.head()
```

```
# STEP 4: Missing Values
print("Missing Values:\n", df.isnull().sum())

# Remove rows with missing 'title' (critical column)
df = df[df['title'].notnull()]

# Fill other missing values with "Unknown" or appropriate default
df['director'].fillna("Unknown", inplace=True)
df['cast'].fillna("Unknown", inplace=True)
df['country'].fillna("Unknown", inplace=True)
df['date_added'].fillna("Unknown", inplace=True)
df['rating'].fillna("Unknown", inplace=True)
```

```
# STEP 5: Remove Duplicates
df.drop_duplicates(inplace=True)
```

```
# STEP 6: Standardize Text Columns
df['type'] = df['type'].str.strip().str.lower()
df['country'] = df['country'].str.strip().str.title()
df['rating'] = df['rating'].str.strip().str.upper()
```

```
# STEP 7: Convert 'date_added' to datetime format
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Optional: Extract year/month
df['added_year'] = df['date_added'].dt.year
df['added_month'] = df['date_added'].dt.month
```

```
# STEP 8: Rename Columns (lowercase, underscores)
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

```
# STEP 9: Final Dataset Check
print("Final Shape:", df.shape)
df.info()
df.head()
```

```
# STEP 9: Final Dataset Check
print("Final Shape:", df.shape)
df.info()
df.head()
```

```
# STEP 10: Export Cleaned Dataset
df.to_csv("netflix_cleaned.csv", index=False)
```
